```python
x=df
# x=df.drop("species",axis=1)
x
```

```python
y=iris["target"]
# y=df["species"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=101)
```

Using Support vector Machine

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(x_train, y_train)
```

```python
prediction=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, prediction))
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
accuracy_score(y_test,prediction)
```

Using Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(x_train,y_train)
```

```python
prediction=rfc.predict(x_test)
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
print(classification_report(y_test,prediction))
```

```python
accuracy_score(y_test, prediction)
```

Using Logistic Regression

```python
prediction=log.predict(x_test)
```

```python
accuracy_score(y_test, prediction)
```

Using K Nearest Neighbour


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
prediction=knn.predict(x_test)
```

```python
accuracy_score(y_test,prediction)
```

we need to check for what value of n_neighbour can give more accurate score


```python
score=[]
for i in range(1,60):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred=knn.predict(x_test)
    score.append(accuracy_score(y_test,pred))
```

```python
plt.plot(range(1,60), score, color="blue", 
         linestyle="dashed", marker="o", markerfacecolor='red')
```

As we can see it gives good accuracy_sscore of 100% at n_neighbor = 1

```python
l=[6.2,
3.4,
5.4,
2.3]
a=np.asarray(l)
data=a.reshape(1,-1)
prediction=knn.predict(data)
if prediction==0:
    print("setosa")
elif prediction==1:
    print("vesicolor")
else:
    print("virginica")
```
